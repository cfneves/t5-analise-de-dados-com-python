## 1. Arquivos: CSV e JSON

### 🔹 Exemplo 1 (resolvido) — Escrevendo e lendo um arquivo JSON

```python
import json

produto = {"nome": "Mouse", "preco": 45.0, "estoque": 12}

with open("produto.json", "w", encoding="utf-8") as arquivo:
    json.dump(produto, arquivo, ensure_ascii=False)

with open("produto.json", "r", encoding="utf-8") as arquivo:
    dados = json.load(arquivo)

print(f"{dados['nome']} custa R$ {dados['preco']}")
```

### 🔹 Exemplo 2 (resolvido) — Escrevendo uma lista de dicionários em CSV

```python
import csv

alunos = [
    {"nome": "Ana", "nota": 8.5},
    {"nome": "Bruno", "nota": 7.0},
    {"nome": "Carla", "nota": 9.0},
]

with open("alunos.csv", "w", newline="", encoding="utf-8") as arquivo:
    escritor = csv.DictWriter(arquivo, fieldnames=["nome", "nota"])
    escritor.writeheader()
    escritor.writerows(alunos)

with open("alunos.csv", "r", encoding="utf-8") as arquivo:
    print(arquivo.read())
```

## 2. Datas e Expressões Regulares

### 🔹 Exemplo 1 (resolvido) — Calculando dias restantes com `datetime`

```python
from datetime import date

hoje = date.today()
fim_do_curso = date(2026, 8, 14)

dias_restantes = (fim_do_curso - hoje).days
print(f"Faltam {dias_restantes} dias para o fim do curso.")
```

### 🔹 Exemplo 2 (resolvido) — Validando um e-mail com `re`

```python
import re

email = "aluno@exemplo.com"
padrao = r".+@.+\..+"

if re.match(padrao, email):
    print("E-mail parece válido!")
else:
    print("E-mail inválido.")
```

## 3. Funções: parâmetros, valores padrão e `return`

### 🔹 Exemplo 1 (resolvido) — Função com retorno

```python
def media(lista):
    return sum(lista) / len(lista)

notas = [7.5, 8.0, 6.5]
print(media(notas))
```

### 🔹 Exemplo 2 (resolvido) — Parâmetro opcional (valor padrão)

```python
def calcular_total(preco, quantidade, desconto=0):
    return preco * quantidade - desconto

print(calcular_total(10, 2))        # sem desconto: 20
print(calcular_total(10, 2, 5))     # com desconto: 15
```

## 4. Lambda e Módulos

### 🔹 Exemplo 1 (resolvido) — Função lambda

```python
media_lambda = lambda lista: sum(lista) / len(lista)

print(media_lambda([7.5, 8.0, 6.5]))
```

### 🔹 Exemplo 2 (resolvido) — Criando e usando um módulo próprio

No Colab, criamos o "módulo" como um arquivo de texto e importamos com `%%writefile` (truque específico do Colab/Jupyter).

```python
%%writefile utilidades.py
def media(lista):
    return sum(lista) / len(lista)

def calcular_total(preco, quantidade, desconto=0):
    return preco * quantidade - desconto
```

```python
import utilidades

print(utilidades.media([7.5, 8.0, 6.5]))
print(utilidades.calcular_total(10, 2, 5))
```
